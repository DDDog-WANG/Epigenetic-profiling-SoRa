# Original data (multi-cells image)

```python
path_home = "/home/acd13264yb/DDDog/Datasets/221206SoRa/"
path_home = "../../Datasets/221206SoRa/"

path_chip = ["Ctrl/60x_4x_Dapi_test_50nmSTP006/60x_4x_Dapi_test_50nmSTP006_XY00",
             "VPA/VPA_555CTCF_488H3K27AC_DAPI_003/VPA_555CTCF_488H3K27AC_DAPI_003_XY00"]
Chip_status = ["H3K27ac", "CTCF", "Hoechst"]
```

```python
path = path_home+path_chip[0]+"1.ome.tif"
print(path)

# img = Image.open(path)
# img = cv2.imread(path,-1)
# img = mpimg.imread(path,-1)
img_uint16 = io.imread(path)
print(img_uint16.shape)
print("dtype: ", img_uint16.dtype)

plt.figure(figsize=(20,20))
plt.subplot(1,4,1)
plt.title("Total max: "+str(img_uint16.max())+", min: "+str(img_uint16.min()))
plt.imshow(img_uint16.astype(np.uint8))
if len(img_uint16.shape)==3:
    for i in range(3):
        plt.subplot(1,4,i+2)
        plt.title(str(Chip_status[i])+" max: "+str(img_uint16.max())+", min: "+str(img_uint16.min()))
        plt.imshow(img_uint16[:,:,i])
```

```python
img_uint8 = img_uint16 - img_uint16.min()
img_uint8 = img_uint8 / (img_uint8.max() - img_uint8.min())
img_uint8 *= 255
img_uint8 = img_uint8.astype(np.uint8)
print(img_uint8.shape)
print("dtype: ", img_uint8.dtype)

plt.figure(figsize=(20,20))
plt.subplot(1,4,1)
plt.title("Total max: "+str(img_uint8.max())+", min: "+str(img_uint8.min()))
plt.imshow(img_uint8)
if len(img_uint8.shape)==3:
    for i in range(3):
        plt.subplot(1,4,i+2)
        plt.title(str(Chip_status[i])+" max: "+str(img_uint8.max())+", min: "+str(img_uint8.min()))
        plt.imshow(img_uint8[:,:,i])
```
